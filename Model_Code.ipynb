Random Forest

```python
# Importing necessary libraries
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pickle
from google.colab import drive

# Assuming x_ros and y_ros are your features and target respectively

# Splitting the data into training and testing sets
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_ros, y_ros, test_size=0.25, random_state=42)

# Initializing and fitting the Random Forest Classifier
dt1 = RandomForestClassifier()
dt1.fit(x_train1, y_train1)

# Evaluating the model
train_score = dt1.score(x_train1, y_train1)
test_score = dt1.score(x_test1, y_test1)
print("Train Score : ", train_score, " Test Score : ", test_score)

# Making predictions
y_pred1 = dt1.predict(x_test1)

# Computing accuracy
accuracy = accuracy_score(y_test1, y_pred1)
print("Accuracy: ", accuracy * 100)

# Computing f1-score
f1 = f1_score(y_test1, y_pred1)
print("F1-score: ", f1)

# Computing recall
recall = recall_score(y_test1, y_pred1)
print("Recall: ", recall)

# Computing precision
precision = precision_score(y_test1, y_pred1)
print("Precision: ", precision)

print(x_train1.shape, y_train1.shape)

# Mount Google Drive
drive.mount('/content/drive')

# Define the file path to save the model
file_path = "/content/drive/My Drive/heart_disease_risk_prediction_model.pkl"

# Save the trained model to Google Drive
with open(file_path, 'wb') as file:
    pickle.dump(dt1, file)
```

```python
from sklearn.metrics import roc_auc_score

# Getting the predicted probabilities for the positive class (class 1)
y_probs = dt1.predict_proba(x_test1)[:, 1]

# Computing the AUC score
auc = roc_auc_score(y_test1, y_probs)
print("AUC Score:", auc)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Getting the probabilities for each class
y_probs = dt1.predict_proba(x_test1)

# Extracting probabilities of the positive class (class 1)
y_probs = y_probs[:, 1]

# Calculating the ROC curve
fpr, tpr, thresholds = roc_curve(y_test1, y_probs)

# Calculating the AUC score
auc = roc_auc_score(y_test1, y_probs)

# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

- **PhysicalHealth**: Numerical value (030) representing the number of days in the past 30 days that the person's physical health was not good.  

- **MentalHealth**: Numerical value (030) representing the number of days in the past 30 days that the person's mental health was not good.  

- **DiffWalking**: Binary variable indicating difficulties in walking  
  - Yes = 1  
  - No = 0  

- **Sex**: Binary variable indicating gender  
  - Male = 1  
  - Female = 0  

- **AgeCategory**: Categorical variable representing age categories  
  - 18-24 = 0

  - 30-34 = 2  
  - 35-39 = 3  
  - 40-44 = 4  
  - 45-49 = 5

  - 60-64 = 8  
  - 65-69 = 9  
  - 70-74 = 10  
  - 75-79 = 11  
  - 80 or older = 12  

- **Race**: Categorical variable representing race  
  - White = 0  
  - Black or African American = 1  
  - Asian = 2  
  - American Indian or Alaska Native = 3  
  - Native Hawaiian or Other Pacific Islander = 4  
  - Multiple/Other = 5  

- **Diabetic**: Binary variable indicating diabetes status  
  - Yes = 1  
  - No = 0  

- **PhysicalActivity**: Continuous numerical value representing physical activity level.  

- **GenHealth**: Categorical variable representing general health condition  
  - Excellent = 0  
  - Very Good = 1  
  - Good = 2  
  - Fair = 3  
  - Poor = 4  

- **SleepTime**: Continuous numerical value representing average hours of sleep.

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Getting the probabilities for each class
y_probs = nb_model.predict_proba(x_test)

# Extracting probabilities of the positive class (class 1)
y_probs = y_probs[:, 1]

# Calculating the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calculating the AUC score
auc = roc_auc_score(y_test, y_probs)

# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (AUC = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
